```python
# Import Useful Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# from the Coingeco Library Import the Coingeco API
from pycoingecko import CoinGeckoAPI
# Create the Coingeco API object as cg
cg = CoinGeckoAPI()
```

```python
#confirm connection
cg.ping()
```

```python
# get a list of coins, sort df by id
coinList = cg.get_coins_list()
```

```python
coinList
```

```python
coinDataFrame = pd.DataFrame.from_dict(coinList).sort_values('id'
                                      ).reset_index(drop=True)
```

```python
coinDataFrame
```

```python
# This returns the currency data and other relivant information
coins = cg.get_coins_markets('usd')
coins
```

```python
# Desired Values from the Data returned
keys = [
    'id', 'symbol', 'name', 'current_price', 'market_cap',
    'market_cap_rank', 'total_volume', 'circulating_supply',
    'total_supply', 'max_supply'
]
```

```python
# Filter the list of dictionaries using list comprehension
filtered_data = [{key: c[key] for key in keys} for c in coins]
```

```python
filtered_data
```

```python
# Convert to DataFrame
df = pd.DataFrame(filtered_data)
```

```python
df.sample(10)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Tomndjerry@localhost:5432/Coindatabase')
```

```python
# Export DataFrame to SQL
df.to_sql('my_table', engine, if_exists='replace', index=False)
```
